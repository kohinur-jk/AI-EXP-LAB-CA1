# Linear regression

```python
import torch
import torch.nn as nn
```

```python
tr.columns # all colums
```

```python
tr_x_col = list(tr.columns)
tr_x_col.remove('price')
tr_y_col = ['price']
```

```python
tr_x_col, tr_y_col
```

```python
tr_x_df = pd.DataFrame(tr, columns=tr_x_col)
tr_y_df = pd.DataFrame(tr, columns=tr_y_col)
```

```python
tr_x = torch.tensor(tr_x_df.values, dtype=torch.float)
tr_y = torch.tensor(tr_y_df.values, dtype=torch.float)
```

```python
tr_x.shape, tr_y.shape
```

# Do it now, use nn.sequential late

```python
class net(nn.Module):
    def __init__(self, inp, hidden1, hidden2, hidden3, output):
        super(net, self).__init__()
        
        self.linear1 = nn.Linear(inp, hidden1)
        self.linear2 = nn.Linear(hidden1, hidden2)
        self.linear3 = nn.Linear(hidden2, hidden3)
        self.linear4 = nn.Linear(hidden3, output)
        
    def forward(self, x):
        y_pred = self.linear1(x).clamp(min=0)
        y_pred = self.linear2(y_pred).clamp(min=0)
        y_pred = self.linear3(y_pred).clamp(min=0)
        y_pred = self.linear4(y_pred)
        
        return y_pred
              
```

```python
h1, h2, h3 = 5000, 15000, 500
```

```python
inp, output = tr_x.shape[1], tr_y.shape[1]
```

# first prediction

```python
model1 = net(inp, h1, h2, h3, output)
```

```python
# criterion

criterion = nn.MSELoss(reduction='sum')
```

```python
optimizer = torch.optim.SGD(model1.parameters(), lr=1e-4)
```

```python
losses1 = []

for t in range(500):
    y_pred = model1(tr_x)
    
    loss = criterion(y_pred, tr_y)
    print(t, loss.item())
    losses1.append(loss.item())
    
    if torch.isnan(loss):
        break
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```
